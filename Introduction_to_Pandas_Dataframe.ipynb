##Membuat DataFrame dari 2D array(Matriks)

##Series with Dictionary

```python
my_series = pd.Series({"Jabar" : "Bandung", "Jateng" : "Semarang", "Jatim" : "Surabaya", "Bali" : "Denpansar"})
my_series
```

```python
df_series = pd.DataFrame(data = my_series.items(),
                         index = range(0,4),
                         columns = ["Provinsi", "Kota"])
df_series
```

```python
#Membuat dataframe untuk menampilkan hasil uji algoritma
df_uji = pd.DataFrame({"Algoritma 1": [88,84,90,30,20],
                       "Algoritma 2": [3,4,5,6,7],
                       "Algoritma 3": [9,10,11,12,13]}, index =['Akurasi', 'Recall', 'Presisi', 'ROC/AUC', 'F1-Score'])
df_uji
```
